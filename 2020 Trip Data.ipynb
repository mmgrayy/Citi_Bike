```python
#Import needed dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
#import all tripdata from 2021
file_1= os.path.join("Data/jan2020.csv" )
file_2 = os.path.join("Data/feb2020.csv" )
file_3 = os.path.join("Data/march2020.csv" )
file_4 = os.path.join("Data/april2020.csv" )
file_5 = os.path.join("Data/may2020.csv" )
file_6 = os.path.join("Data/june2020.csv" )
file_7 = os.path.join("Data/july2020.csv" )
file_8 = os.path.join("Data/aug2020.csv" )
file_9 = os.path.join("Data/sep2020.csv" )
file_10 = os.path.join("Data/oct2020.csv" )
file_11 = os.path.join("Data/nov2020.csv" )
file_12= os.path.join("Data/dec2020.csv" )

jan = pd.read_csv(file_1)
feb = pd.read_csv(file_2)
march = pd.read_csv(file_3)
april = pd.read_csv(file_4)
may = pd.read_csv(file_5)
june = pd.read_csv(file_6)
july = pd.read_csv(file_7)
aug = pd.read_csv(file_8)
sep = pd.read_csv(file_9)
october = pd.read_csv(file_10)
nov = pd.read_csv(file_11)
dec = pd.read_csv(file_12)
```

```python
#merge tabels
frames = [jan, feb, march, april,
           may, june, july, aug,
           sep, october, nov, dec]
```

```python
entire_year = pd.concat(frames)
```

```python
entire_year.head()
```

```python
entire_year.to_csv("Trip History 2020.csv")
```
